```python
import cv2
import os

# Daftar 10 gambar
images_list = ['img_1.jpg', 'img_2.jpg', 'img_3.jpg', 'img_4.jpg', 'img_5.jpg',
               'img_6.jpg', 'img_7.jpg', 'img_8.jpg', 'img_9.jpg', 'img_10.jpg']

if not os.path.exists('output_results'):
    os.makedirs('output_results')

for img_name in images_list:
    img = cv2.imread(img_name)
    if img is not None:
        # Manipulasi dengan dimensi yang berbeda
        res = cv2.resize(img, (500, 500))

        # hasil dalam format .jpg
        output_name = f"output_results/resized_{img_name}"
        cv2.imwrite(output_name, res)
        print(f"Berhasil memproses {img_name}")
    else:
        print(f"Gagal membaca {img_name}")
```
